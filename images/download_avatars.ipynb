```python
import requests # request img from web
import shutil # save img locally
import pandas as pd
import shutil
```

```python
roster = pd.read_csv (r'roster_names.csv')
```

```python
roster['formatted_name'] = roster['name'].replace(r'[^\w\s]+', '').replace('â€™', '', regex=True).replace(' ', '-', regex=True)
```

```python
roster['url'] = 'https://www.2kratings.com/wp-content/uploads/' + roster['formatted_name'] + "-2K-Rating.png"
```

```python
roster['url2'] = 'https://www.2kratings.com/wp-content/uploads/' + roster['formatted_name'] + "-2K.png"
```

```python
#roster
```

```python
no_image_players = []

def save_image(url, url2, file_name, name):
  tried_all = False
  src = url
  urls_tried = 0
  while(urls_tried < 2):
  
    res = requests.get(src, stream = True)

    if res.status_code == 200:
        with open(file_name,'wb') as f:
            shutil.copyfileobj(res.raw, f)
        print('Image sucessfully Downloaded: ',file_name)
        return
    else:
        src = url2
        urls_tried += 1
  
  print('Image Couldn\'t be retrieved', file_name)
  no_image_players.append(name)
```

```python
# Loop through urls

for index, row in roster.iterrows():
  url = row['url']
  url2 = row['url2']
  file_name = "./images/" + row['formatted_name'] + ".png"
  name = row['name']
  #print(row['name'], row['url'])
  save_image(url, url2, file_name, name)
```
